# Algoritmos de Ordenamiento

Los **algoritmos de ordenamiento** reorganizan los elementos de una lista de menor a mayor.
En esta libreta exploramos tres algoritmos clásicos:

| Algoritmo   | Comparaciones  | Intercambios máx. | Espacio extra | Estable |
|-------------|:--------------:|:-----------------:|:-------------:|:-------:|
| Burbuja     | O(n²)          | O(n²)             | O(1)          | Sí      |
| Selección   | O(n²)          | **O(n)**          | O(1)          | No      |
| Merge Sort  | O(n log n)     | —                 | **O(n)**      | Sí      |

> **Burbuja y Selección** son simples de implementar pero lentos para listas grandes (n > 10 000).  
> **Merge Sort** usa la estrategia *divide y vencerás* y garantiza O(n log n) en **todos los casos**,
> a costa de memoria auxiliar O(n).

---
## 1. Ordenamiento de Burbuja (*Bubble Sort*)

### Idea
Recorre la lista comparando **pares adyacentes** e intercambiándolos si están en el orden
incorrecto. El elemento más grande "burbujea" hacia el final en cada pasada.
Se repite hasta que no queden intercambios pendientes.

### Pseudocódigo

```
BubbleSort(lista):
    n <- longitud(lista)
    i <- 0
    
    para i hasta n-1 hacer
        intercambiado <- falso
        
        para j <- 0 hasta n-2-i hacer
            si A[j] > A[j+1] entonces
                intercambiar(A[j], A[j+1])
                intercambiado <- verdadero
                
        si intercambiado == falso entonces
            terminar    // optimizacion: lista ya ordenada
```

### Complejidad

| Caso     | Tiempo  | Intercambios | Nota                           |
|----------|---------|:------------:|--------------------------------|
| Peor     | O(n²)   | O(n²)        | Lista completamente invertida  |
| Promedio | O(n²)   | O(n²)        |                                |
| Mejor    | **O(n)**| 0            | Lista ya ordenada (con flag)   |
| Espacio  | O(1)    | —            | Ordenamiento *in-place*        |


````python
def burbuja(lista):
    '''
    Ordenamiento de burbuja — O(n^2).
    Devuelve una nueva lista ordenada de menor a mayor sin modificar la original.
    '''
    arr = lista[:]          # trabajar sobre una copia
    n = len(arr)
    for i in range(n - 1):
        intercambiado = False
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                intercambiado = True
        if not intercambiado:
            break           # optimizacion: ningún intercambio -> ya ordenado
    return arr


# ── Ejemplo rapido ────────────────────────────────────────────────────────────
datos = [64, 34, 25, 12, 22, 11, 90]
print('Original:', datos)
print('Ordenada:', burbuja(datos))
````

### Traza paso a paso
La siguiente función muestra el estado de la lista al **terminar cada pasada**
y cuenta comparaciones e intercambios totales.


````python
def burbuja_traza(lista):
    arr = lista[:]
    n = len(arr)
    total_comp = 0
    total_swap = 0

    print('Lista inicial:', arr)
    print()

    for i in range(n - 1):
        swaps_pasada = 0
        intercambiado = False
        for j in range(n - 1 - i):
            total_comp += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                intercambiado = True
                swaps_pasada += 1
                total_swap += 1
        etiqueta = 'intercambio' if swaps_pasada == 1 else 'intercambios'
        print(f'Pasada {i + 1:2d}: {arr}   ({swaps_pasada} {etiqueta})')
        if not intercambiado:
            print('          ^ sin cambios — lista ya ordenada')
            break

    print()
    print('Resultado:    ', arr)
    print('Comparaciones:', total_comp)
    print('Intercambios: ', total_swap)
    return arr


print('=== Caso tipico ===')
burbuja_traza([64, 34, 25, 12, 22, 11, 90])

print()
print('=== Mejor caso: lista ya ordenada ===')
burbuja_traza([1, 2, 3, 4, 5])

print()
print('=== Peor caso: lista invertida ===')
burbuja_traza([5, 4, 3, 2, 1])
````

## 2. Ordenamiento por Selección (*Selection Sort*)

### Idea
Divide la lista en dos partes: **ordenada** (izquierda) y **sin ordenar** (derecha).
En cada iteración busca el **mínimo** de la parte sin ordenar y lo coloca
al inicio de esa parte con un único intercambio.

### Pseudocódigo

```
SelectionSort(lista):
    i <- 0
    n <- longitud(lista)
    
    para i  hasta n-1 hacer
        min_idx <- i
        
        para j <- i+1 hasta n-1 hacer
            si A[j] < A[min_idx] entonces
                min_idx <- j
                
        si min_idx != i entonces
            intercambiar(A[i], A[min_idx])
```

### Complejidad

| Caso        | Tiempo  | Intercambios | Nota                                        |
|-------------|---------|:------------:|---------------------------------------------|
| Peor        | O(n²)   | O(n)         | Siempre hace n(n-1)/2 comparaciones         |
| Promedio    | O(n²)   | O(n)         | No hay atajo posible                        |
| Mejor       | O(n²)   | 0            | Lista ya ordenada (pero igual compara todo) |
| Espacio     | O(1)    | —            | Ordenamiento *in-place*                     |

**Ventaja clave:** a lo sumo **n − 1 intercambios** — útil cuando intercambiar es costoso
(p. ej. escribir en disco o memoria lenta).


````python
def seleccion(lista):
    '''
    Ordenamiento por seleccion — O(n^2) comparaciones, O(n) intercambios.
    Devuelve una nueva lista ordenada de menor a mayor sin modificar la original.
    '''
    arr = lista[:]          # trabajar sobre una copia
    n = len(arr)
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr


# ── Ejemplo rapido ────────────────────────────────────────────────────────────
datos = [64, 34, 25, 12, 22, 11, 90]
print('Original:', datos)
print('Ordenada:', seleccion(datos))
````

### Traza paso a paso
Muestra cada iteración: el mínimo encontrado, si hubo intercambio y el estado de la lista.


````python
def seleccion_traza(lista):
    arr = lista[:]
    n = len(arr)
    total_comp = 0
    total_swap = 0

    print('Lista inicial:', arr)
    print()

    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            total_comp += 1
            if arr[j] < arr[min_idx]:
                min_idx = j

        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            total_swap += 1
            msg = f'Iter {i+1:2d}: min={arr[i]:3d} (estaba en [{min_idx}]) -> intercambiar [{i}]<->[{min_idx}]'
        else:
            msg = f'Iter {i+1:2d}: min={arr[i]:3d} (ya en [{i}])              -> sin intercambio   '

        print(f'{msg}   {arr}')

    print()
    print('Resultado:    ', arr)
    print('Comparaciones:', total_comp)
    print('Intercambios: ', total_swap)
    return arr


print('=== Caso tipico ===')
seleccion_traza([64, 34, 25, 12, 22, 11, 90])

print()
print('=== Lista ya ordenada ===')
seleccion_traza([1, 2, 3, 4, 5])
````

## Intermedio: Recursividad

Antes de estudiar Merge Sort es necesario entender **recursividad**, porque el algoritmo
se llama a sí mismo para resolver versiones más pequeñas del problema.

---

### ¿Qué es una función recursiva?

Una función es **recursiva** cuando se llama a sí misma dentro de su propia definición.
Para que no se llame infinitamente, toda función recursiva debe tener dos partes:

```
funcion_recursiva(problema):
    si el problema es lo suficientemente pequeño:   ← CASO BASE
        resolver directamente y retornar

    reducir el problema en una versión más pequeña
    llamar a funcion_recursiva(problema_más_pequeño) ← LLAMADA RECURSIVA
    combinar el resultado y retornar
```

| Parte             | ¿Qué hace?                                              |
|-------------------|---------------------------------------------------------|
| **Caso base**     | Detiene la recursión. Sin él, la función se llama sin fin. |
| **Caso recursivo**| Reduce el problema y delega el resto a la misma función. |


### ¿Cuándo usar recursividad?

Es especialmente útil cuando el problema tiene una estructura naturalmente recursiva, como:

- Árboles (recorrer nodos)

- Sistemas de archivos

- Búsqueda binaria

- Divide y vencerás (Merge Sort, Quick Sort)
---

### La pila de llamadas (*call stack*)

Cada vez que una función se llama a sí misma, Python "apila" una nueva ejecución
encima de la anterior. Cuando llega al caso base, empieza a "desapilar" y a retornar
resultados hacia atrás:

```
factorial(4)
  └─ 4 * factorial(3)
           └─ 3 * factorial(2)
                    └─ 2 * factorial(1)
                             └─ caso base → retorna 1
                    retorna 2 * 1 = 2
           retorna 3 * 2 = 6
  retorna 4 * 6 = 24
```

> **Importante:** cada nivel de recursión tiene sus propias variables locales.
> No comparten memoria entre sí.

---

### Ejemplo 1 — Cuenta regresiva

El problema más simple: contar de `n` hasta 0.

```
cuenta_regresiva(n):
    si n < 0 entonces retornar      ← caso base

    imprimir n
    cuenta_regresiva(n - 1)         ← llamada recursiva con problema más pequeño
```

---

### Ejemplo 2 — Factorial

El factorial de n se define como `n! = n × (n-1) × (n-2) × … × 1`,
con el caso base `0! = 1`.

Expresado recursivamente: **n! = n × (n−1)!**

```
factorial(n):
    si n == 0 entonces retornar 1   ← caso base

    retornar n * factorial(n - 1)   ← llamada recursiva
```

---

### Ejemplo 3 — Suma de una lista

Sumar todos los elementos de una lista también puede verse de forma recursiva:
la suma de la lista completa es el **primer elemento** más la **suma del resto**.

```
suma(lista):
    si lista está vacía entonces retornar 0         ← caso base

    retornar lista[0] + suma(lista[1:])             ← llamada recursiva
```

---

### ¿Por qué importa para Merge Sort?

Merge Sort divide la lista en dos mitades y **se llama a sí mismo** sobre cada mitad.
El caso base es una lista de 1 elemento (ya está ordenada). La parte de "combinar"
es la operación de fusión (*merge*).

Sin entender recursividad, el código de Merge Sort parece magia.
Con ella, cada llamada resuelve un problema más pequeño hasta llegar al caso base.

````python
# ── Ejemplo 1: Cuenta regresiva ───────────────────────────────────────────────

def cuenta_regresiva(n):
    if n < 0:               # caso base: detener cuando llegamos a -1
        return
    print(n, end='  ')
    cuenta_regresiva(n - 1) # llamada recursiva: problema más pequeño

print('Cuenta regresiva desde 5:')
cuenta_regresiva(5)
print()

# ── Ejemplo 2: Factorial ───────────────────────────────────────────────────────

def factorial(n):
    if n == 0:              # caso base: 0! = 1
        return 1
    return n * factorial(n - 1)   # n! = n × (n-1)!

print('\nFactorial:')
for k in range(8):
    print(f'  {k}! = {factorial(k)}')

# ── Ejemplo 3: Suma de lista ───────────────────────────────────────────────────

def suma(lista):
    if len(lista) == 0:     # caso base: lista vacía → suma 0
        return 0
    return lista[0] + suma(lista[1:])   # primer elemento + suma del resto

print('\nSuma recursiva:')
numeros = [3, 7, 2, 9, 4]
print(f'  lista = {numeros}')
print(f'  suma  = {suma(numeros)}  (verificación con sum(): {sum(numeros)})')

# ── Comparar recursión vs iteración ───────────────────────────────────────────

print('\nFactorial recursivo vs iterativo:')

def factorial_iterativo(n):
    resultado = 1
    for i in range(2, n + 1):
        resultado *= i
    return resultado

for k in [0, 1, 5, 10]:
    rec = factorial(k)
    ite = factorial_iterativo(k)
    ok  = '✓' if rec == ite else '✗'
    print(f'  {k:2d}! = {rec:10d}   iterativo = {ite:10d}   {ok}')
````

## 3. Merge Sort (*Ordenamiento por Fusión*)

### Idea

Aplica la estrategia **divide y vencerás** en tres pasos:

1. **Dividir** — partir la lista por la mitad hasta llegar a sublistas de 1 elemento
   (una lista de 1 elemento ya está ordenada por definición).
2. **Vencer** — ordenar recursivamente cada mitad.
3. **Combinar** — **fusionar** (*merge*) las dos mitades ordenadas en una sola lista ordenada.

```
lista original        [64, 34, 25, 12, 22, 11, 90]
                       ↙                          ↘
              [64, 34, 25]               [12, 22, 11, 90]
               ↙        ↘                ↙              ↘
          [64]       [34, 25]       [12, 22]         [11, 90]
                      ↙    ↘         ↙    ↘           ↙    ↘
                   [34]   [25]    [12]   [22]       [11]   [90]
                      ↘    ↙         ↘    ↙           ↘    ↙
                   [25, 34]        [12, 22]          [11, 90]
               ↘        ↙                ↘              ↙
           [25, 34, 64]               [11, 12, 22, 90]
                       ↘                          ↙
               [11, 12, 22, 25, 34, 64, 90]   ← resultado
```

### Pseudocódigo

```
MergeSort(lista, inicio, fin):
    si inicio >= fin entonces
        retornar                           // caso base: 1 elemento ya está ordenado

    medio <- (inicio + fin) / 2

    MergeSort(lista, inicio, medio)        // ordenar mitad izquierda
    MergeSort(lista, medio + 1, fin)       // ordenar mitad derecha
    Mezclar(lista, inicio, medio, fin)     // fusionar ambas mitades ordenadas


Mezclar(lista, inicio, medio, fin):
    izq <- lista[inicio .. medio]          // copias temporales de cada mitad
    der <- lista[medio+1 .. fin]

    i <- 0,  j <- 0,  k <- inicio

    mientras i < |izq| y j < |der| hacer
        si izq[i] <= der[j] entonces
            lista[k] <- izq[i]
            i <- i + 1
        si no
            lista[k] <- der[j]
            j <- j + 1
        k <- k + 1

    // copiar elementos restantes (solo uno de los dos bloques tendrá restos)
    mientras i < |izq| hacer
        lista[k] <- izq[i];  i <- i + 1;  k <- k + 1
    mientras j < |der| hacer
        lista[k] <- der[j];  j <- j + 1;  k <- k + 1
```

### Complejidad

| Caso     | Tiempo       | Espacio extra | Nota                                           |
|----------|:------------:|:-------------:|------------------------------------------------|
| Peor     | O(n log n)   | O(n)          | Siempre divide y fusiona de forma equilibrada  |
| Promedio | O(n log n)   | O(n)          | No depende del orden inicial de la lista       |
| Mejor    | O(n log n)   | O(n)          | No hay atajo: siempre divide hasta el fondo    |

**Ventajas:**
- Garantía de O(n log n) en **cualquier** caso de entrada.
- **Estable**: elementos iguales mantienen su orden relativo original.
- Eficiente para listas enlazadas y datos externos (disco).

**Desventaja:** necesita O(n) de memoria extra para los arreglos auxiliares.

````python
def _mezclar(izq, der):
    '''
    Fusiona dos listas ordenadas en una nueva lista ordenada.
    Complejidad: O(|izq| + |der|) tiempo y espacio.
    '''
    resultado = []
    i = j = 0
    while i < len(izq) and j < len(der):
        if izq[i] <= der[j]:      # <= garantiza estabilidad
            resultado.append(izq[i])
            i += 1
        else:
            resultado.append(der[j])
            j += 1
    resultado.extend(izq[i:])     # copiar restos (solo uno tendrá elementos)
    resultado.extend(der[j:])
    return resultado


def mergesort(lista):
    '''
    Merge sort (fusión) — O(n log n) en todos los casos.
    Divide la lista en mitades, ordena cada una recursivamente
    y las fusiona. Devuelve una nueva lista sin modificar la original.
    '''
    if len(lista) <= 1:
        return lista[:]                      # caso base

    medio = len(lista) // 2
    izq   = mergesort(lista[:medio])         # ordenar mitad izquierda
    der   = mergesort(lista[medio:])         # ordenar mitad derecha
    return _mezclar(izq, der)                # combinar


# ── Ejemplo rápido ────────────────────────────────────────────────────────────
datos = [64, 34, 25, 12, 22, 11, 90, 1, 8, 3]
print('Original:', datos)
print('Ordenada:', mergesort(datos))
````

### Traza paso a paso

La siguiente función muestra el **árbol de recursión** completo: cada nivel de
indentación representa un nivel más profundo de la recursión.  
La flecha `↓ dividir` indica la llamada recursiva; `↑ mezclar` indica el retorno con el resultado fusionado.

````python
def mergesort_traza(lista, prof=0):
    '''
    Merge sort con traza del árbol de recursión.
    prof  — profundidad actual (usada internamente para la indentación).
    '''
    sangria = '  ' * prof

    if len(lista) <= 1:
        print(f'{sangria}base    : {lista}')
        return lista[:]

    medio = len(lista) // 2
    print(f'{sangria}↓ dividir: {lista}  →  izq={lista[:medio]}  der={lista[medio:]}')

    izq = mergesort_traza(lista[:medio], prof + 1)
    der = mergesort_traza(lista[medio:], prof + 1)

    resultado = _mezclar(izq, der)
    print(f'{sangria}↑ mezclar: {izq} + {der}  →  {resultado}')
    return resultado


# ── Traza completa ────────────────────────────────────────────────────────────
datos = [64, 34, 25, 12, 22, 11, 90]
print('Lista inicial:', datos)
print()
resultado = mergesort_traza(datos)
print()
print('Resultado:', resultado)
````

## 4. Comparación directa

| Métrica           | Burbuja  | Selección | Merge Sort   |
|-------------------|:--------:|:---------:|:------------:|
| Comparaciones     | O(n²)    | O(n²)     | O(n log n)   |
| Intercambios/mov. | O(n²)    | **O(n)**  | O(n log n)   |
| Espacio extra     | O(1)     | O(1)      | O(n)         |
| Estable           | Sí       | No        | Sí           |

La siguiente celda mide los tres algoritmos sobre la misma lista aleatoria y
compara comparaciones y movimientos reales con la teoría.

````python
import random
import math


def contar_burbuja(lista):
    arr = lista[:]
    n, comp, swap = len(arr), 0, 0
    for i in range(n - 1):
        intercambiado = False
        for j in range(n - 1 - i):
            comp += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                intercambiado = True
                swap += 1
        if not intercambiado:
            break
    return comp, swap


def contar_seleccion(lista):
    arr = lista[:]
    n, comp, swap = len(arr), 0, 0
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            comp += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            swap += 1
    return comp, swap


def contar_mergesort(lista):
    comp_total = [0]
    mov_total  = [0]

    def _merge_contar(izq, der):
        resultado = []
        i = j = 0
        while i < len(izq) and j < len(der):
            comp_total[0] += 1
            if izq[i] <= der[j]:
                resultado.append(izq[i]); i += 1
            else:
                resultado.append(der[j]); j += 1
            mov_total[0] += 1
        resto = izq[i:] + der[j:]
        mov_total[0] += len(resto)
        resultado.extend(resto)
        return resultado

    def _ms(lst):
        if len(lst) <= 1:
            return lst[:]
        m = len(lst) // 2
        return _merge_contar(_ms(lst[:m]), _ms(lst[m:]))

    _ms(lista)
    return comp_total[0], mov_total[0]


# ── Comparar sobre la misma lista aleatoria ───────────────────────────────────
random.seed(42)
n = 20
datos = random.sample(range(1, 100), n)

cb, sb  = contar_burbuja(datos)
cs, ss  = contar_seleccion(datos)
cm, mm  = contar_mergesort(datos)

teoria_n2      = n * (n - 1) // 2
teoria_nlogn   = round(n * math.log2(n))

print('Lista (n=%d): %s' % (n, datos))
print()
print('%-22s %15s %17s' % ('Algoritmo', 'Comparaciones', 'Intercambios/mov.'))
print('-' * 56)
print('%-22s %15d %17d' % ('Burbuja',     cb, sb))
print('%-22s %15d %17d' % ('Seleccion',   cs, ss))
print('%-22s %15d %17d' % ('Merge Sort',  cm, mm))
print()
print('Referencia teórica:')
print('  O(n²)      → n*(n-1)/2 = %d' % teoria_n2)
print('  O(n log n) → n*log2(n) ≈ %d' % teoria_nlogn)
````
