## 2. Variables y tipos de datos

Python es de **tipado dinámico**: no necesitas declarar el tipo de una variable. El intérprete lo deduce automáticamente en tiempo de ejecución.

| Tipo | Ejemplo | Descripción |
|------|---------|-------------|
| `int` | `42` | Número entero |
| `float` | `3.14` | Número decimal |
| `str` | `"hola"` | Cadena de texto |
| `bool` | `True` / `False` | Booleano |
| `NoneType` | `None` | Ausencia de valor |

````python
# Declaración de variables — el tipo se asigna automáticamente
entero   = 42
decimal  = 3.14159
texto    = "Hola Python"
booleano = True
nulo     = None

# type() nos dice el tipo de una variable
print(type(entero))    # <class 'int'>
print(type(decimal))   # <class 'float'>
print(type(texto))     # <class 'str'>
print(type(booleano))  # <class 'bool'>
print(type(nulo))      # <class 'NoneType'>
````

````python
# Python permite reasignar una variable con un tipo distinto (tipado dinámico)
x = 10
print(f"x = {x}, tipo: {type(x).__name__}")

x = "ahora soy un string"
print(f"x = {x}, tipo: {type(x).__name__}")

# Asignación múltiple en una línea
a, b, c = 1, 2, 3
print(f"\na={a}, b={b}, c={c}")

# Intercambiar dos variables — ¡sin variable auxiliar!
a, b = b, a
print(f"Tras swap: a={a}, b={b}")
````

## 3. Operaciones aritméticas

Python soporta todas las operaciones matemáticas básicas. Nota que `/` siempre devuelve un `float`, mientras que `//` hace división entera.

````python
x, y = 17, 5

print(f"{x} + {y}  = {x + y}")      # suma           → 22
print(f"{x} - {y}  = {x - y}")      # resta          → 12
print(f"{x} * {y}  = {x * y}")      # multiplicación → 85
print(f"{x} / {y}  = {x / y}")      # división real  → 3.4  (siempre float)
print(f"{x} // {y} = {x // y}")     # división entera→ 3    (floor)
print(f"{x} % {y}  = {x % y}")      # módulo/residuo → 2
print(f"{x} ** {y} = {x ** y}")     # potencia       → 1419857
````

````python
import math 
````

````python
math.pi
````

````python
print(f"sqrt(144)   = {math.sqrt(144)}")     # raíz cuadrada → 12.0
print(f"pi          = {math.pi:.5f}")        # π con 6 decimales
print(f"e           = {math.e:.6f}")         # número de Euler
print(f"ceil(3.2)   = {math.ceil(3.2)}")     # techo → 4
print(f"floor(3.9)  = {math.floor(3.9)}")    # piso  → 3
print(f"abs(-7)     = {abs(-7)}")            # valor absoluto → 7
print(f"round(3.567, 2) = {round(3.567, 2)}") # redondeo → 3.57
````

## 4. Strings

Los strings en Python son **inmutables** (no se pueden modificar en el lugar). Los métodos devuelven un nuevo string.

Las **f-strings** (desde Python 3.6) son la forma moderna y recomendada de formatear texto.

````python
nombre = "    Python básico  "

print(f"Original : '{nombre}'")
print(f"strip()  : '{nombre.strip()}'")          # elimina espacios extremos
print(f"upper()  : '{nombre.strip().upper()}'")  # a mayúsculas
print(f"lower()  : '{nombre.strip().lower()}'")  # a minúsculas
print(f"title()  : '{nombre.strip().title()}'")  # primera letra de cada palabra
print(f"len()    : {len(nombre.strip())}")       # longitud
````

````python
# f-strings — la forma más limpia de insertar variables en texto
lenguaje = "Python"
version  = 3.12
pi       = 3.14159265

print(f"Lenguaje: {lenguaje}, Versión: {version}")
print(f"Pi con 3 decimales: {pi:.3f}")
print(f"Número alineado a la derecha (5 dígitos): {42:5d}")
print(f"Porcentaje: {0.867:.1%}")
````

````python
# split, join, replace y búsqueda
frase = "la,programacion,en,python,es,divertida"

palabras = frase.split(",")          # divide por separador → lista
print(type(palabras))
print("split(',')  :", palabras)

unida = " ".join(palabras)           # une lista con separador → string
print("join(' ')   :", unida)

cambiada = unida.replace("python", "Python")
print("replace()   :", cambiada)

# Verificar contenido
print(f"\n'Python' in frase: {'Python' in cambiada}")
print(f"Empieza con 'la'  : {cambiada.startswith('la')}")
print(f"Termina con 'ida' : {cambiada.endswith('ida')}")
````

## 5. Listas

Las listas son la estructura de datos más usada en Python. Son **mutables**, **ordenadas** y pueden contener **tipos mixtos**.

```python
mi_lista = [elemento0, elemento1, elemento2, ...]
```

El índice comienza en `0`. Los índices negativos cuentan desde el final (`-1` es el último).

````python
"python"[::-1]
````

````python
# Crear una lista
frutas = ["manzana", "naranja", "uva", "mango", "kiwi"]


print("Lista completa :", frutas)
print("Índice 0       :", frutas[0])     # primer elemento
print("Índice -1      :", frutas[-1])    # último elemento
print("Slice [1:3]    :", frutas[1:3])   # elementos en posición 1 y 2
print("Slice [::2]    :", frutas[::2])   # cada 2 elementos (paso 2)
print("Slice [::-1]   :", frutas[::-1])  # invertida
````

````python
# Modificar la lista
frutas = ["manzana", "naranja", "uva"]
print("Original:", frutas)

frutas.append("mango")          # agrega al final
frutas.insert(1, "kiwi")        # inserta en posición 1
print("append + insert:", frutas)

frutas.remove("uva")            # elimina la primera ocurrencia de "uva"
ultimo = frutas.pop()           # elimina y devuelve el último
print(f"remove + pop (sacado='{ultimo}'):", frutas)

frutas.sort()                   # ordena en el lugar (modifica la lista)
print("sort():", frutas)
````

````python
# Operaciones estadísticas básicas
notas = [78, 92, 85, 60, 97, 73, 88]

print(f"Notas     : {notas}")
print(f"Cantidad  : {len(notas)}")
print(f"Mínima    : {min(notas)}")
print(f"Máxima    : {max(notas)}")
print(f"Suma      : {sum(notas)}")
print(f"Promedio  : {sum(notas)/len(notas):.2f}")
print(f"Ordenadas : {sorted(notas)}")
````

````python
# List comprehension — forma compacta de crear listas a partir de otras

# [expresión for elemento in iterable if condición]

cuadrados = [x**2 for x in range(1, 8)]
print("Cuadrados [1..7]:", cuadrados)

pares = [x for x in range(1, 11) if x % 2 == 0]
print("Pares [1..10]   :", pares)

# Convertir temperatura de Celsius a Fahrenheit
celsius    = [0, 20, 37, 100]
fahrenheit = [(c * 9/5) + 32 for c in celsius]
print(f"Celsius    : {celsius}")
print(f"Fahrenheit : {fahrenheit}")
````

## 6. Diccionarios

Los diccionarios almacenan pares **clave → valor**. Son equivalentes a los *hash maps* en otros lenguajes.

- Las **claves** son únicas y normalmente strings o números.
- Los **valores** pueden ser cualquier tipo, incluyendo listas u otros diccionarios.
- Son **mutables** y desde Python 3.7 mantienen el orden de inserción.

````python
# Crear un diccionario
estudiante = {
    "nombre" : "Ana López",
    "carnet" : 202300001,
    "notas"  : [85, 90, 78],
    "activo" : True,
}

# Acceder a valores
print("Nombre :", estudiante["nombre"])               # acceso directo
print("Carnet :", estudiante.get("carnet"))           # .get() es más seguro
print("Curso  :", estudiante.get("curso", "Sin asignar"))  # valor por defecto
````

````python
# Modificar y agregar entradas
estudiante["email"] = "ana@usac.edu.gt"    # nueva clave
estudiante["activo"] = False               # modificar existente
estudiante["notas"].append(92)             # modificar lista dentro del dict

print("Diccionario actualizado:")
for clave, valor in estudiante.items():
    print(f"  {clave:10} : {valor}")

# Eliminar una clave
del estudiante["activo"]
print(f"\nTras del 'activo' — claves: {list(estudiante.keys())}")
````

````python
# Diccionario de diccionarios — estructura más compleja
cursos = {
    "F12": {"nombre": "Programación",       "creditos": 4, "semestre": 1},
    "M11": {"nombre": "Cálculo I",           "creditos": 5, "semestre": 1},
    "F21": {"nombre": "Física Clásica I",    "creditos": 4, "semestre": 2},
}

print("Cursos disponibles:")
for codigo, info in cursos.items():
    print(f"  [{codigo}] {info['nombre']:25} | {info['creditos']} créditos | Semestre {info['semestre']}")
````

## 7. Condicionales

Python usa indentación (espacios o tabulaciones) para definir bloques de código — **no hay llaves `{}`**.

```python
if condicion1:
    # bloque 1
elif condicion2:
    # bloque 2
else:
    # bloque por defecto
```

````python
# Clasificación de notas
def clasificar_nota(nota):
    if nota >= 90:
        return "Excelente (A)"
    elif nota >= 80:
        return "Muy Bueno (B)"
    elif nota >= 70:
        return "Bueno (C)"
    elif nota >= 60:
        return "Suficiente (D)"
    else:
        return "Reprobado (F)"

for nota in [95, 83, 71, 60, 45]:
    print(f"  Nota {nota} → {clasificar_nota(nota)}")
````

````python
# Operadores de comparación
a, b = 10, 20

print(f"{a} == {b} : {a == b}")   # igual
print(f"{a} != {b} : {a != b}")   # diferente
print(f"{a} <  {b} : {a <  b}")   # menor que
print(f"{a} >  {b} : {a >  b}")   # mayor que
print(f"{a} <= {b} : {a <= b}")   # menor o igual
print(f"{a} >= {b} : {a >= b}")   # mayor o igual

# Operadores lógicos
x = 15
print(f"\n{x} > 10 and {x} < 20: {x > 10 and x < 20}")
print(f"{x} < 10 or  {x} > 5 : {x < 10 or  x > 5}")
print(f"not ({x} > 10)        : {not (x > 10)}")

# Forma pythónica de verificar rangos
print(f"\n10 < {x} < 20: {10 < x < 20}")
````

````python
# Operador ternario (condicional en una línea)
# valor_si_verdadero if condicion else valor_si_falso

edad = 20
estado = "mayor de edad" if edad >= 18 else "menor de edad"
print(f"Edad {edad}: {estado}")

# in — verificar pertenencia
vocales = ['a', 'e', 'i', 'o', 'u']
for letra in ['a', 'b', 'e', 'z']:
    print(f"'{letra}' es vocal: {'Sí' if letra in vocales else 'No'}")
````

## 8. Ciclos

Python tiene dos tipos de ciclos:
- `for`: itera sobre cualquier **iterable** (lista, string, range, etc.)
- `while`: repite mientras una **condición** sea verdadera

Palabras clave especiales:
- `break`: termina el ciclo inmediatamente
- `continue`: salta a la siguiente iteración
- `else` en ciclos: se ejecuta si el ciclo terminó sin `break`

````python
# for sobre una lista
lenguajes = ["Python", "C++", "Java", "JavaScript"]

for lenguaje in lenguajes:
    print(f"  → {lenguaje}")
````

````python
# range(inicio, fin, paso) — genera secuencia de números
print("range(5)        :", list(range(5)))          # 0..4
print("range(1, 6)     :", list(range(1, 6)))       # 1..5
print("range(0, 10, 2) :", list(range(0, 10, 2)))   # 0,2,4,6,8
print("range(10, 0, -2):", list(range(10, 0, -2)))  # cuenta regresiva

print("\nTabla del 3:")
for i in range(1, 11):
    print(f"  3 × {i} = {3*i}")
````

````python
# enumerate() — índice + valor al mismo tiempo
frutas = ["manzana", "naranja", "uva"]
print("enumerate():")
for i, fruta in enumerate(frutas, start=1):
    print(f"  [{i}] {fruta}")

# zip() — iterar dos listas al mismo tiempo
nombres = ["Ana", "Luis", "María"]
notas   = [90, 78, 95]
print("\nzip():")
for nombre, nota in zip(nombres, notas):
    print(f"  {nombre}: {nota}")
````

````python
# while — se usa cuando no se conoce el número de iteraciones
print("Cuenta regresiva:")
cuenta = 5
while cuenta > 0:
    print(f"  {cuenta}...")
    cuenta -= 1
print("  ¡Despegue!")

# break y continue
print("\nbreak — buscar el primer par:")
numeros = [3, 7, 1, 4, 9, 2, 8]
for n in numeros:
    if n % 2 == 0:
        print(f"  Primer par encontrado: {n}")
        break

print("\ncontinue — saltar impares:")
for n in range(1, 8):
    if n % 2 != 0:
        continue  # salta al siguiente si es impar
    print(f"  Par: {n}")
````

## 9. Funciones

Las funciones permiten reutilizar código. En Python se definen con `def`.

```python
def nombre_funcion(param1, param2=valor_default):
    """Docstring: descripción de la función."""
    # cuerpo de la función
    return resultado
```

**Principio:** una función debería hacer **una sola cosa** y hacerla bien.

````python
# Función básica con docstring
def saludar(nombre, saludo="¡Hola"):
    """Genera un saludo personalizado.
    
    Args:
        nombre (str): nombre de la persona a saludar
        saludo (str): tipo de saludo (por defecto '¡Hola')
    
    Returns:
        str: el mensaje de saludo
    """
    return f"{saludo}, {nombre}!"

print(saludar("Ana"))                        # usa saludo por defecto
print(saludar("Luis", "Buenos días"))        # saludo personalizado
print(saludar(saludo="Buenas noches", nombre="María"))  # por nombre (kwargs)
````

````python
# Función que retorna múltiples valores
def estadisticas(numeros):
    """Calcula estadísticas básicas de una lista.
    
    Returns:
        tuple: (mínimo, máximo, promedio)
    """
    return min(numeros), max(numeros), sum(numeros) / len(numeros)

datos = [10, 25, 7, 18, 42, 33, 15]
minimo, maximo, promedio = estadisticas(datos)  # desempaquetado de tupla

print(f"Datos    : {datos}")
print(f"Mínimo   : {minimo}")
print(f"Máximo   : {maximo}")
print(f"Promedio : {promedio:.2f}")
````

````python
# Función recursiva — se llama a sí misma
def factorial(n):
    """Calcula n! de forma recursiva.
    
    factorial(0) = 1
    factorial(n) = n × factorial(n-1)
    """
    if n <= 1:           # caso base — detiene la recursión
        return 1
    return n * factorial(n - 1)  # caso recursivo

for n in range(8):
    print(f"  {n}! = {factorial(n)}")
````

````python
# Función lambda — función anónima de una línea
# lambda argumentos: expresión

cuadrado = lambda x: x ** 2
suma     = lambda a, b: a + b

print(f"cuadrado(7)  = {cuadrado(7)}")
print(f"suma(3, 4)   = {suma(3, 4)}")

# Las lambdas son útiles como argumentos de sorted(), map(), filter()
alumnos = [("Ana", 90), ("Luis", 78), ("María", 95), ("Carlos", 83)]

# Ordenar por nota (segundo elemento de la tupla)
por_nota = sorted(alumnos, key=lambda alumno: alumno[1], reverse=True)
print("\nRanking por nota:")
for i, (nombre, nota) in enumerate(por_nota, start=1):
    print(f"  {i}. {nombre}: {nota}")
````

## 10. Manejo de errores

Los errores en Python se llaman **excepciones**. El bloque `try/except` permite capturarlas y manejarlas sin que el programa se detenga.

```python
try:
    # código que puede fallar
except TipoDeError as e:
    # qué hacer si ocurre ese error
else:
    # se ejecuta si NO hubo error
finally:
    # se ejecuta SIEMPRE (con o sin error)
```

Errores comunes:
| Excepción | Causa |
|-----------|-------|
| `ZeroDivisionError` | División entre cero |
| `TypeError` | Tipo incorrecto en una operación |
| `ValueError` | Valor inválido (ej: `int("abc")`) |
| `IndexError` | Índice fuera de rango |
| `KeyError` | Clave inexistente en diccionario |
| `FileNotFoundError` | Archivo no encontrado |

````python
# Ejemplo básico de try/except
def dividir(a, b):
    """Divide a entre b con manejo de errores."""
    try:
        resultado = a / b
    except ZeroDivisionError:
        print(f"  ✗ Error: no se puede dividir {a} entre 0")
        return None
    except TypeError as e:
        print(f"  ✗ TypeError: {e}")
        return None
    else:
        print(f"  ✓ {a} / {b} = {resultado}")
        return resultado
    finally:
        print(f"  [intento de dividir {a!r} / {b!r} completado]")

dividir(10, 2)
print()
dividir(10, 0)
print()
dividir("diez", 2)
````

````python
# Capturar múltiples excepciones
def acceder_lista(lista, indice):
    """Intenta acceder a lista[indice] con manejo de errores."""
    try:
        return lista[indice]
    except IndexError:
        print(f"  ✗ Índice {indice} fuera de rango (tamaño={len(lista)})")
    except TypeError:
        print(f"  ✗ El índice debe ser un entero, recibido: {type(indice).__name__}")

numeros = [10, 20, 30]
print(acceder_lista(numeros, 1))      # OK
acceder_lista(numeros, 10)            # IndexError
acceder_lista(numeros, "uno")         # TypeError
````

````python
import math
````

````python
# raise — lanzar una excepción manualmente

def raiz_cuadrada(n):
    """Calcula la raíz cuadrada. Rechaza números negativos con un mensaje claro."""
    if not isinstance(n, (int, float)):
        raise TypeError(f"Se esperaba int o float, se recibió: {type(n).__name__}")
    if n < 0:
        raise ValueError(f"No se puede calcular la raíz de un número negativo: {n}")
    return math.sqrt(n)

for valor in [25, 0, -9, "hola"]:
    try:
        resultado = raiz_cuadrada(valor)
        print(f"  sqrt({valor}) = {resultado}")
    except (ValueError, TypeError) as e:
        print(f"  ✗ Error con {valor!r}: {e}")
````
