# Búsqueda de 2 en 2 con Retroceso

Este notebook implementa un algoritmo de búsqueda sobre listas **ordenadas** con la siguiente idea:

1. Avanzar de **2 en 2** posiciones: revisar índices 0, 2, 4, 6, ...
2. Si el elemento actual es **mayor** que el objetivo → **retroceder un paso** y revisar el elemento anterior.
3. Si el elemento actual es **igual** al objetivo → encontrado.
4. Si el elemento actual es **menor** → seguir avanzando.

Es una versión simplificada del **Jump Search** con paso fijo de 2.

| Algoritmo | Requiere orden | Complejidad |
|---|---|---|
| Búsqueda de 2 en 2 con retroceso | Sí (lista ordenada) | $O(n)$ |

## Pseudocódigo general

```
FUNCIÓN busqueda_2en2(lista, objetivo):
    n ← longitud(lista)
    i ← 0

    MIENTRAS i < n:
        SI lista[i] == objetivo:
            RETORNAR i          // encontrado

        SI lista[i] > objetivo:
            SI i > 0 Y lista[i-1] == objetivo:
                RETORNAR i - 1  // retroceder un paso
            RETORNAR -1         // no está en la lista

        i ← i + 2              // avanzar de 2 en 2

    // Revisar el último elemento si fue saltado (n par)
    SI n > 0 Y lista[n-1] == objetivo:
        RETORNAR n - 1

    RETORNAR -1
```

> **¿Por qué funciona el retroceso?**  
> Como la lista está **ordenada**, si `lista[i] > objetivo`, el único lugar donde podría estar el objetivo es `lista[i-1]` (el elemento que saltamos). No puede estar antes porque habríamos detectado el sobrepasamieto antes.

````python
# Utilidades para mostrar resultados

def mostrar_resultado(lista, objetivo, indice):
    """Imprime la lista con el objetivo marcado, o 'no encontrado'."""
    if indice == -1:
        print(f"Lista   : {lista}")
        print(f"Buscar  : {objetivo}  →  NO ENCONTRADO")
    else:
        marcada = list(map(str, lista))
        marcada[indice] = f"[{lista[indice]}]"
        print(f"Lista   : {marcada}")
        print(f"Buscar  : {objetivo}  →  índice {indice}")
    print()
````

## Implementación 1: ciclo `for`

Usamos `range(0, n, 2)` para generar los índices 0, 2, 4, ... directamente.  
Si en algún índice `i` el elemento supera al objetivo, retrocedemos manualmente a `i - 1`.

**Pseudocódigo:**
```
FUNCIÓN busqueda_2en2_for(lista, objetivo):
    n ← longitud(lista)

    PARA i ← 0, 2, 4, ..., n-1 (paso 2):
        SI lista[i] == objetivo:
            RETORNAR i

        SI lista[i] > objetivo:
            SI i > 0 Y lista[i-1] == objetivo:
                RETORNAR i - 1
            RETORNAR -1

    // Último elemento (puede haber sido saltado si n es par)
    SI n es par Y lista[n-1] == objetivo:
        RETORNAR n - 1

    RETORNAR -1
```

````python
def busqueda_2en2_for(lista, objetivo):
    """
    Búsqueda de 2 en 2 con retroceso usando ciclo FOR.
    Requiere lista ordenada de menor a mayor.
    Retorna el índice del objetivo, o -1 si no se encuentra.
    """
    n = len(lista)

    for i in range(0, n, 2):          # avanzar: 0, 2, 4, ...
        if lista[i] == objetivo:
            return i                   # encontrado exactamente

        if lista[i] > objetivo:        # sobrepasamos el objetivo
            if i > 0 and lista[i - 1] == objetivo:
                return i - 1           # retroceder un paso
            return -1                  # no estaba en la lista

    # Si n es par, el último índice (n-1) fue saltado por el for
    if n % 2 == 0 and n > 0 and lista[n - 1] == objetivo:
        return n - 1

    return -1
````

````python
print("=== Ejemplos con ciclo FOR ===")

lista = [1, 3, 5, 7, 9, 11, 13, 15]

# Caso 1: objetivo en índice par
mostrar_resultado(lista, 7,  busqueda_2en2_for(lista, 7))

# Caso 2: objetivo en índice impar (requiere retroceso)
mostrar_resultado(lista, 3,  busqueda_2en2_for(lista, 3))

# Caso 3: objetivo no está en la lista
mostrar_resultado(lista, 6,  busqueda_2en2_for(lista, 6))

# Caso 4: primer elemento
mostrar_resultado(lista, 1,  busqueda_2en2_for(lista, 1))

# Caso 5: último elemento (índice impar → necesita retroceso)
mostrar_resultado(lista, 15, busqueda_2en2_for(lista, 15))

# Caso 6: lista con largo impar
lista_impar = [2, 4, 6, 8, 10]
mostrar_resultado(lista_impar, 10, busqueda_2en2_for(lista_impar, 10))
````

## Implementación 2: ciclo `while`

Controlamos el índice manualmente con `i += 2`. Esto permite más flexibilidad, por ejemplo para modificar el paso en tiempo de ejecución.

**Pseudocódigo:**
```
FUNCIÓN busqueda_2en2_while(lista, objetivo):
    n ← longitud(lista)
    i ← 0

    MIENTRAS i < n:
        SI lista[i] == objetivo:
            RETORNAR i

        SI lista[i] > objetivo:
            SI i > 0 Y lista[i-1] == objetivo:
                RETORNAR i - 1
            RETORNAR -1

        i ← i + 2

    // Verificar el último elemento si fue saltado
    SI lista[n-1] == objetivo:
        RETORNAR n - 1

    RETORNAR -1
```

````python
def busqueda_2en2_while(lista, objetivo):
    """
    Búsqueda de 2 en 2 con retroceso usando ciclo WHILE.
    Requiere lista ordenada de menor a mayor.
    Retorna el índice del objetivo, o -1 si no se encuentra.
    """
    n = len(lista)
    i = 0

    while i < n:
        if lista[i] == objetivo:
            return i                       # encontrado

        if lista[i] > objetivo:            # sobrepasamos
            if i > 0 and lista[i - 1] == objetivo:
                return i - 1               # retroceder
            return -1

        i += 2                             # avanzar de 2 en 2

    # Salimos del while; el último elemento pudo haberse saltado
    if n > 0 and lista[n - 1] == objetivo:
        return n - 1

    return -1
````

````python
print("=== Ejemplos con ciclo WHILE ===")

lista = [2, 5, 8, 12, 17, 21, 30, 35, 42]

# Caso 1: objetivo en índice par
mostrar_resultado(lista, 8,  busqueda_2en2_while(lista, 8))

# Caso 2: objetivo en índice impar (retroceso)
mostrar_resultado(lista, 5,  busqueda_2en2_while(lista, 5))

# Caso 3: objetivo mayor a todos los elementos
mostrar_resultado(lista, 50, busqueda_2en2_while(lista, 50))

# Caso 4: objetivo menor a todos los elementos
mostrar_resultado(lista, 1,  busqueda_2en2_while(lista, 1))

# Caso 5: lista de un solo elemento encontrado
mostrar_resultado([7], 7,    busqueda_2en2_while([7], 7))

# Caso 6: lista de un solo elemento no encontrado
mostrar_resultado([7], 3,    busqueda_2en2_while([7], 3))
````

## Implementación 3: versión verbose (paso a paso)

Igual que la versión `while`, pero imprime cada paso del algoritmo para visualizar su recorrido. Útil para depuración y comprensión.

**Pseudocódigo:**
```
FUNCIÓN busqueda_2en2_verbose(lista, objetivo):
    n ← longitud(lista)
    i ← 0
    pasos ← 0

    IMPRIMIR "Inicio: lista =", lista, "| objetivo =", objetivo

    MIENTRAS i < n:
        pasos ← pasos + 1
        IMPRIMIR "Paso", pasos, "→ comparando índice", i, "valor", lista[i]

        SI lista[i] == objetivo:
            IMPRIMIR "Encontrado en índice", i
            RETORNAR i

        SI lista[i] > objetivo:
            IMPRIMIR "Sobrepasamos: lista[", i, "]=", lista[i], ">" , objetivo
            SI i > 0:
                IMPRIMIR "Retroceder → revisar índice", i-1, "valor", lista[i-1]
                SI lista[i-1] == objetivo:
                    RETORNAR i - 1
            IMPRIMIR "No encontrado"
            RETORNAR -1

        i ← i + 2

    SI lista[n-1] == objetivo:
        RETORNAR n - 1

    RETORNAR -1
```

````python
def busqueda_2en2_verbose(lista, objetivo):
    """
    Búsqueda de 2 en 2 con retroceso — versión VERBOSE.
    Imprime cada paso del algoritmo para visualizar el recorrido.
    """
    n = len(lista)
    i = 0
    paso = 0

    print(f"Lista   : {lista}")
    print(f"Objetivo: {objetivo}")
    print("-" * 40)

    while i < n:
        paso += 1
        print(f"Paso {paso:2d} | i={i:2d} | lista[{i}]={lista[i]:3d}", end="  ")

        if lista[i] == objetivo:
            print(f"== {objetivo}  →  ENCONTRADO")
            return i

        if lista[i] > objetivo:
            print(f"> {objetivo}   →  SOBREPASAMOS")
            if i > 0:
                paso += 1
                print(f"Paso {paso:2d} | i={i-1:2d} | lista[{i-1}]={lista[i-1]:3d}", end="  ")
                print(f"retroceso", end="  ")
                if lista[i - 1] == objetivo:
                    print(f"== {objetivo}  →  ENCONTRADO")
                    return i - 1
                else:
                    print(f"!= {objetivo}  →  no está")
            print("Resultado: -1 (NO ENCONTRADO)")
            return -1

        print(f"< {objetivo}   →  avanzar")
        i += 2

    # Verificar último elemento si fue saltado
    if n > 0 and lista[n - 1] == objetivo:
        print(f"Último elemento lista[{n-1}]={lista[n-1]}  →  ENCONTRADO")
        return n - 1

    print("Resultado: -1 (NO ENCONTRADO)")
    return -1
````

````python
print("=" * 50)
print("EJEMPLO A: objetivo en índice impar (retroceso)")
print("=" * 50)
lista = [1, 3, 5, 7, 9, 11, 13, 15]
busqueda_2en2_verbose(lista, 3)
````

````python
print("=" * 50)
print("EJEMPLO B: objetivo en índice par (sin retroceso)")
print("=" * 50)
busqueda_2en2_verbose(lista, 9)
````

````python
print("=" * 50)
print("EJEMPLO C: objetivo NO está en la lista")
print("=" * 50)
busqueda_2en2_verbose(lista, 6)
````

## Casos borde y pruebas automáticas

Verificamos que las tres implementaciones sean equivalentes en distintos escenarios.

````python
import random

def busqueda_lineal(lista, objetivo):
    """Referencia: búsqueda lineal clásica O(n)."""
    for i, x in enumerate(lista):
        if x == objetivo:
            return i
    return -1


def prueba_equivalencia(n_pruebas=500, seed=42):
    """Compara busqueda_2en2_for y while contra búsqueda lineal."""
    rng = random.Random(seed)
    errores = 0

    for _ in range(n_pruebas):
        tamaño = rng.randint(0, 30)
        lista = sorted(rng.randint(0, 50) for _ in range(tamaño))
        objetivo = rng.randint(0, 50)

        ref = busqueda_lineal(lista, objetivo)
        # La referencia puede dar un índice; si hay duplicados, cualquier índice válido sirve
        esperado = lista[ref] if ref != -1 else None

        for fn in [busqueda_2en2_for, busqueda_2en2_while]:
            idx = fn(lista, objetivo)
            obtenido = lista[idx] if idx != -1 else None

            if obtenido != esperado:
                print(f"ERROR en {fn.__name__}")
                print(f"  lista={lista}, objetivo={objetivo}")
                print(f"  esperado índice={ref} (valor={esperado})")
                print(f"  obtenido índice={idx} (valor={obtenido})")
                errores += 1

    if errores == 0:
        print(f"Todas las {n_pruebas} pruebas pasaron correctamente.")
    else:
        print(f"{errores} errores encontrados.")


prueba_equivalencia()
````

## Análisis de complejidad

| Situación | Pasos máximos | Razón |
|---|---|---|
| Objetivo en índice par | $\frac{n}{2}$ comparaciones | Avanzamos hasta encontrarlo |
| Objetivo en índice impar | $\frac{n}{2} + 1$ comparaciones | Avanzamos y retrocedemos una vez |
| No está en la lista | $\frac{n}{2}$ comparaciones | Llegamos al final o sobrepasamos |

En todos los casos el número de comparaciones es proporcional a $\frac{n}{2}$, que en notación Big O se simplifica a:

$$T(n) = O\left(\frac{n}{2}\right) = O(n)$$

> El factor constante $\frac{1}{2}$ se descarta en Big O. Este algoritmo es más rápido que la búsqueda lineal en la **práctica** (revisa ~la mitad de los elementos), pero **asintóticamente** sigue siendo $O(n)$.
